  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
def f(x):
  return 10*x - 5
```

```python
import matplotlib.pyplot as plt
import numpy as np

xs = np.linspace(0, 10, 10)
plt.plot(xs, f(xs), 'o');
```

```python
import sympy as sp
x = sp.Symbol('x')

f = 10 * x - 5
derivative_f = sp.diff(f, x)
print("Похідна функції 10x-5:", derivative_f)
```

```python
# Визначення функції та її похідної
x = np.linspace(0, 10, 100)
f = 10 * x - 5
df = np.full_like(x, 10)  # Похідна f'(x) = 10 (постійна)

# Точка дотику
x_point = 2
f_point = 10 * x_point - 5
df_point = 10
tangent_line = df_point * (x - x_point) + f_point  # Рівняння дотичної

# Побудова графіку
plt.figure(figsize=(10, 5))
plt.plot(x, f, label='f(x) = 10x - 5', color='blue')
plt.plot(x, df, label="f'(x) = 10", color='green')

# Додання точки та дотичної
plt.scatter(x_point, f_point, color='red')  # Точка на кривій
plt.plot(x, tangent_line, '--', color='red', label='Tangent at x = 2')  # Дотична

# Оформлення графіку
plt.title("Графік функції f(x) = 10x - 5 та її похідної")
plt.xlabel("x (години навчання)")
plt.ylabel("y (оцінка)")
plt.legend()
plt.grid(True)

plt.show()
```

# До цього моменту я робила сама, далі все GPTchat , тому що я просто не розумію як його побудувати. І застрягла на цьому

```python
# Крок 1: Символьна змінна і функції
x = sp.Symbol('x')
f_sin = sp.sin(x)
df_sin = sp.diff(f_sin, x)

f_log = sp.ln(x)
df_log = sp.diff(f_log, x)

# Крок 2: Перетворення на числові функції
f_sin_func = sp.lambdify(x, f_sin, 'numpy')
df_sin_func = sp.lambdify(x, df_sin, 'numpy')

f_log_func = sp.lambdify(x, f_log, 'numpy')
df_log_func = sp.lambdify(x, df_log, 'numpy')

# Крок 3: Масив значень x
x_vals = np.linspace(0.1, 10, 200)  # уникаємо 0 для ln(x)

# Крок 4: Обчислення значень функцій
y_sin = f_sin_func(x_vals)
y_sin_prime = df_sin_func(x_vals)

y_log = f_log_func(x_vals)
y_log_prime = df_log_func(x_vals)

# Крок 5: Побудова графіків
plt.figure(figsize=(14, 6))

# Перша підділянка: sin(x)
plt.subplot(1, 2, 1)
plt.plot(x_vals, y_sin, label='f(x) = sin(x)', color='blue')
plt.plot(x_vals, y_sin_prime, label="f'(x) = cos(x)", color='orange', linestyle='--')
plt.title("Синус і його похідна")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

# Друга підділянка: ln(x)
plt.subplot(1, 2, 2)
plt.plot(x_vals, y_log, label='f(x) = ln(x)', color='green')
plt.plot(x_vals, y_log_prime, label="f'(x) = 1/x", color='red', linestyle='--')
plt.title("Логарифм і його похідна")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

𝑓(𝑥)=𝑥2−5𝑥+1

```python
x = sp.symbols('x')
f = x**2 - x*5 + 1
df = sp.diff(f, x)
print("Функція:", f)
print("Похідна функції x**2 - x*5 + 1:", df)
```

```python
f_func = sp.lambdify(x, f, modules='numpy')
f_prime_func = sp.lambdify(x, df, modules='numpy')
```

```python
x = np.linspace(0, 100, 100)
```

```python
f_vals = f_func(x)
f_prime_vals = f_prime_func(x)
```

```python
plt.figure(figsize=(12, 6))
plt.plot(x, f_vals, label='f(x) = x² - 5x + 1', color='blue')
plt.plot(x, f_prime_vals, label="f'(x) = 2x - 5", color='green')

x_point = 5
f_point = x_point**2 - x_point*5 + 1
df_point = 2 * x_point - 5
tangent_line = df_point * (x - x_point) + f_point  # Рівняння дотичної

plt.scatter(x_point, f_point, color='red') # точка на кривій
plt.plot(x, tangent_line, '--', color='red', label='Tangent at x = 5') # дотична

plt.title("Графік функції та її похідної")
plt.xlabel("x (кількість одиниць продукції)")
plt.ylabel("Витрати")
plt.legend()
plt.grid(True)

plt.show()
```
