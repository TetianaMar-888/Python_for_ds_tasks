**2. Базові операції**

Створіть Numpy-масив зі значеннями [12, 24, 36, 48, 60].

Виконайте наступні операції над **оригінальним** масивом і запишіть у нову змінну результат (кожен пункт оцінюється в 0.5 бала):
- Помножте всі елементи на 2
- Додайте до всіх елементів 10
- Піднесіть всі елементи до квадрату
- Обчисліть квадратний корінь з усіх елементів

Виведіть результат виконання кожної операції.

```python
arr = np.array([12, 24, 36, 48, 60])
arr_mul = arr * 2 #Помножте всі елементи на 2
print(arr_mul)
arr_add = arr + 10 #Додайте до всіх елементів 10
print(arr_add)
arr_sq = arr ** 2 #Піднесіть всі елементи до квадрату
print(arr_sq)
arr_sqrt = np.sqrt(arr) #Обчисліть квадратний корінь з усіх елементів
print(arr_sqrt)
```

**3. Статистичні функції**

Створіть Numpy-масив `data` зі значеннями [18, 27, 33, 41, 52, 68, 79].

**4. Індексація**

Створіть Numpy-масив `arr` зі значеннями [15, 22, 35, 44, 58, 67].

Виведіть (кожен пункт оцінюється в 0.5 бала):
- Перший елемент
- Останній елемент
- Елементи з індексами 1 до 3
- Елементи більші за 40

```python
arr = np.array([15, 22, 35, 44, 58, 67])
print(arr[0]) # Перший елемент
print(arr[-1]) # Останній елемент
print(arr[1:4]) # Елементи з індексами від 1 до 3
mask = arr > 40
print(arr[mask])# Елементи більші за 40
```

**5. Двовимірні масиви**

Задано матрицю 3x3.

Виконайте наступні дії з функціоналом Numpy (кожен пункт оцінюється в 0.5 бала):
- Виведіть форму (shape) матриці
- Виведіть перший рядок
- Виведіть другу колонку
- Обчисліть суму всіх елементів
- Обчисліть середнє всіх елементів

```python
matrix = np.array([[2, 4, 6],
                   [8, 10, 12],
                   [14, 16, 18]])
```

```python
print(matrix.shape) # Форма матриці 3 рядки, 3 колонки
print(matrix[0]) # Перший рядок
print(matrix[:, 1]) # Друга колонка
print(np.sum(matrix)) # Сума всіх елементів
print(np.mean(matrix)) # Середнє всіх елементів
```
