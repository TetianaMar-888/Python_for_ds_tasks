**1. Гістограма розподілу рахунків**
Завантажте датасет Tips (Чайові) з файла у матеріалах `tips.csv`.
Створіть гістограму для колонки `total_bill` з 20 bins.
Опціонально можна додати підписи осей ОХ, ОУ, а також назву графіку.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
df = pd.read_csv('tips.csv')
```

```python
fig = px.histogram(df, x='total_bill', nbins=20,
                 title='Розподіл рахунків',
                 labels={'total_bill': 'Рахунок', 'count': 'Кількість рахунків'},
                 color_discrete_sequence=['#8a2be2'])
fig.show()
```

---

**2. Box plot чайових за днями тижня**

Створіть box plot, який показує розподіл чайових (`tip`) за днями тижня (`day`).
В який день навище медіанне (середнє) значення чайоивх? Зелена горизонтальна лінія всередині кожного box-plot.


```python
fig = px.box(df, x='day', y='tip', color='day',
             title='Розподіл чайових за днями тижня',
             labels={'tip': 'Чайові', 'day': 'День тижня'},
             category_orders={'day': ['Thur', 'Fri', 'Sat', 'Sun']})
fig.show()
```

З графіка бачимо, що найвище середнє значення чайових в неділю 3.15 dol, потім  в  п'ятницю 3.0 dol, суботу 2.75 dol і найнижче в четвер 2.305 dol.
Додала в графік ще category_orders, тому що дні виводились розкидані, не зручно дивитись.

---

**3. Scatter plot: рахунок vs чайові**

Створіть діаграму розсіювання для `total_bill` та `tip`.

```python
df.plot.scatter(x='total_bill', y='tip', alpha=0.5, color='purple')
plt.title('Рахунок та чайові')
plt.xlabel('Рахунок')
plt.ylabel('Чайові')
plt.show()
```

---

**4. Bar chart: середній рахунок за часом**

Створіть стовпчасту діаграму середнього рахунку за часом дня (`time`).

```python
df.groupby('time')['total_bill'].mean().plot(kind='bar', color='green')
plt.title('Середній рахунок за часом')
plt.xlabel('Час дня')
plt.ylabel('Середній рахунок')
plt.show()
```

---

**5. Інтерактивна діаграма розсіювання**

Створіть з допомогою бібліотек Plotly інтерактивний scatter plot для `total_bill` vs `tip` з кольором за статтю (`sex`).


```python
fig = px.scatter(df, x='total_bill', y='tip', color='sex',
                title='Рахунок vs чайові',
                labels={'total_bill': 'Рахунок', 'tip': 'Чайові'})
fig.show()
```

---

**6. Інтерактивний box plot: чайові за днями та статтю**

Створіть інтерактивний box plot для чайових за днями тижня з розбивкою за кольором по статі.

```python
fig = px.box(df, x='day', y='tip', color='sex',
             title='Розподіл чайових за днями тижня',
             labels={'tip': 'Чайові', 'day': 'День тижня', 'sex': 'Стать'},
             category_orders={'day': ['Thur', 'Fri', 'Sat', 'Sun']})
fig.show()
```

Якщо з 2 завдання додати для цікавості ще розподіл за статтю, то ми бачимо, що чоловіки частіше лишають чайові і вищі в ціннику. Але середній показник по чайовим в п'ятницю та неділю вищий у жінок.
