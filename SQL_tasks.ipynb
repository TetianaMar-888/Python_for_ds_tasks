#4. Finding the change in an employee's salary over the last 2 years
Return the difference between the current salary of the employee with emp_no = 10001 and his salary 2 years ago. Response format: employee number, salary_2_years_ago, current salary, difference in salaries 2 years ago and now.

*Знаходження зміни зарплати співробітника протягом останніх 2 років

Виведіть різницю між поточною зарплатою співробітника з emp_no = 10001 та його зарплатою 2 роки тому. Формат відповіді: номер співробітника, salary_2_years_ago, поточна ЗП, різниця у зарплатах 2 роки тому і зараз.

```python
df_diff = pd.read_sql_query("""
WITH current_salary AS (
    SELECT
        emp_no,
        salary,
        from_date
    FROM salaries
    WHERE to_date = '9999-01-01' AND emp_no = 10001
),
salary_2_years_ago AS (
    SELECT
        s.emp_no,
        s.salary,
        s.from_date,
        ROW_NUMBER() OVER (PARTITION BY s.emp_no ORDER BY s.from_date DESC) AS rn
    FROM salaries s
    JOIN current_salary c ON s.emp_no = c.emp_no
    WHERE s.from_date <= DATE_SUB(c.from_date, INTERVAL 2 YEAR)
)
SELECT
    c.emp_no,
    s.salary AS salary_2_years_ago,
    c.salary AS current_salary,
    c.salary - s.salary AS difference
FROM current_salary c
JOIN salary_2_years_ago s
    ON c.emp_no = s.emp_no AND s.rn = 1
WHERE c.emp_no = 10001;
""", con)

df_diff
```

#5. Determining the highest salary increase
Find the employee who received the highest salary increase during the year ending 1 year before the most recent date in the database (maximum value of from_date in the salaries table). Response format: employee number, salary from the previous year, salary grade, and amount of salary change.

*Визначення найвищого зростання зарплати
Знайдіть співробітника, який отримав найбільше збільшення зарплати протягом року, який закінчився 1 рік до найновішої дати в базі даних (максимальне значення from_date в таблиці salaries). Формат відповіді: номер співробітника, ЗП з попереднього року, поточка ЗП, на яку суму змінилась ЗП.

```python
df_highest_salary = pd.read_sql_query("""
WITH max_date AS (
    SELECT MAX(from_date) AS max_date
    FROM salaries
),
prev_year_date AS (
    SELECT DATE_SUB(max_date, INTERVAL 1 YEAR) AS target_date
    FROM max_date
),
filtered AS (
    SELECT s.*
    FROM salaries s
    JOIN prev_year_date p
      ON s.from_date >= p.target_date AND s.from_date <= (SELECT max_date FROM max_date)
),
with_lead AS (
    SELECT
        f.emp_no,
        f.salary,
        LEAD(f.salary) OVER (PARTITION BY f.emp_no ORDER BY f.from_date) AS next_salary
    FROM filtered f
)
SELECT
    emp_no,
    salary AS salary_from_previous_year,
    next_salary AS current_salary,
    next_salary - salary AS salary_change
FROM with_lead
WHERE next_salary IS NOT NULL
ORDER BY salary_change DESC
LIMIT 1;
""", con)

df_highest_salary
```

#6. Average salary and first employee in each position
For each position, output its name, the average salary of employees in it, the name of the first employee (or employees, if several people were hired for this position on the same day) who was hired for this position (according to the hire_date), and the hire_date of this employee or employees.In the response, sort the results by job title and employee name.

*Середня зарплата та перший праціник на кожній посаді

Для кожної посади виведіть її назву, середню зарплату співробітників на ній, ім'я першого співробітника (або співробітників, якщо в один і той самий день було прийнято на цю посаду кілька людей), який був прийнятий на цю посаду (згідно дати прийому на роботу - hire_date), а також hire_date цього співробітника чи співробітників. У відповіді відсортуйте результати за назвою посади і імʼям співробітника.

```python
df_avg = pd.read_sql_query("""
WITH avg_salary AS (
    SELECT
        t.title,
        ROUND(AVG(s.salary), 2) AS avg_salary
    FROM titles t
    JOIN salaries s ON t.emp_no = s.emp_no
    GROUP BY t.title
),
first_hire_date AS (
    SELECT
        t.title,
        MIN(e.hire_date) AS min_hire_date
    FROM titles t
    JOIN employees e ON t.emp_no = e.emp_no
    GROUP BY t.title
),
first_employee AS (
    SELECT
        t.title,
        e.first_name,
        e.last_name,
        e.hire_date
    FROM titles t
    JOIN employees e ON t.emp_no = e.emp_no
    JOIN first_hire_date fh
      ON fh.title = t.title AND fh.min_hire_date = e.hire_date
)
SELECT
    a.title,
    a.avg_salary,
    f.first_name,
    f.last_name,
    f.hire_date
FROM avg_salary a
JOIN first_employee f ON a.title = f.title
ORDER BY a.title, f.first_name, f.last_name;
""", con)

df_avg
```

`emp_no` співробітників, зарплати яких будемо досліджувати: `419748, 496734, 264134, 209076, 86631, 456171, 16887, 230255, 246465, 420097, 44362, 280370, 479975, 433320, 473949`

```python
df_ntile = pd.read_sql_query("""
WITH salary_groups AS (
    SELECT
        e.emp_no,
        e.first_name,
        e.last_name,
        s.salary,
        NTILE(3) OVER (ORDER BY s.salary DESC) AS salary_group
    FROM employees e
    JOIN salaries s ON e.emp_no = s.emp_no
    WHERE e.emp_no IN (419748, 496734, 264134, 209076, 86631, 456171, 16887, 230255, 246465, 420097, 44362, 280370, 479975, 433320, 473949) AND s.to_date = '9999-01-01'
)
SELECT
    emp_no,
    first_name,
    last_name,
    salary,
    salary_group
FROM salary_groups
ORDER BY salary_group, emp_no;
""", con)

df_ntile
```
