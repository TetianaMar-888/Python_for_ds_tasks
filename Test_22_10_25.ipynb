```python
import numpy as np
import pandas as pd

pd.set_option('display.max.rows',130)
pd.set_option('display.max.columns',130)
pd.set_option('float_format', '{:.2f}'.format)
```

```python
URL = "https://people.sc.fsu.edu/~jburkardt/data/csv/cities.csv"
df = pd.read_csv(URL)
```

```python
df.info()
```

```python
df.head()
```

```python
df.columns = df.columns.str.replace('"', '').str.strip()
df['City'] = df['City'].astype(str)
```

```python
df.shape
```

```python
df['City'] = df['City'].str.replace('"', '').str.replace('"', '')
```

```python
df['City']
```

```python
help(str.lower)
```

```python
letters = set("".join(df['City'].str.lower()))
print(letters)
```

```python
#df
```

```python
sorted_letters = sorted(letters)
print(sorted_letters)
```

```python
sorted_letters[3:]
```

```python
df['letters'] = [sorted_letters] * len(df)
```

```python
df["hash32"] = df["letters"].apply(
    lambda s: hash(tuple(s)) % 2**32
)
```

```python
print(df["letters"])
```

#    .      AI  

```python
df["hash32"] = df['letters'].apply(
    lambda s: np.bitwise_or.reduce(
        np.left_shift(
            np.uint32(1),
            np.unique(np.frombuffer("".join(sorted(s)).encode("ascii"), dtype=np.uint8) - ord("a"))
        )
    ).astype(np.uint32)
)
```
